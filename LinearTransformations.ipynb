```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# points a, b and, c
a, b, c, d, e, f, g, h, i, j, k, l, m, n, o = (2, 2), (1.6, 1), (1.4, 1), (1, 2), (1.2, 2), (1.5, 1.2), (1.8, 2), (2, 2), (2, 1), (3, 1), (3, 2), (2.8, 2), (2.8, 1.2), (2.2, 1.2), (2.2, 2)

# matrix with row vectors of points
A = np.array([a, b, c, d, e, f, g, h, i, j, k, l, m, n, o])

## plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = row
    x, y, = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.show()
```

# DILATIONS AND CONTRACTIONS

```python
# create the scaling transformation matrix
# optionally set dilation/contraction factor (c = ?)
c = 2
L = np.array([[c, 0], [0, c]])

# plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = np.matmul(L, row)
    x, y, = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.show()
```

# REFLECTIONS ABOUT AN AXIS

#### X-AXIS

```python
# create the x axis reflection transformation matrix
x = [[1, 0], [0, -1]]
L = np.array(x)

# plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = np.matmul(L, row)
    x, y, = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.show()
```

# ROTATIONS

#### 45 Degree

```python
# create the 45 degree transformation matrix
# optionally: set custom rotation angle here (theta = ?)
theta = np.pi / 4
L = np.array([[np.cos(theta), -(np.sin(theta))], [np.sin(theta), np.cos(theta)]])

# plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = np.matmul(L, row)
    x, y, = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.show()
```

#### 180 Degree

```python
# create the 180 degree transformation matrix
theta = np.pi
L = np.array([[np.cos(theta), -(np.sin(theta))], [np.sin(theta), np.cos(theta)]])

# plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = np.matmul(L, row)
    x, y, = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.show()
```

# TRANSLATIONS USING HOMEOGENOUS COORDINATES

```python
# points a, b and, c
p, q, r, s, t, u, v, w, x, y, z, aa, ab, ac, ad = (2, 2, 1), (1.6, 1, 1), (1.4, 1, 1), (1, 2, 1), (1.2, 2, 1), (1.5, 1.2, 1), (1.8, 2, 1), (2, 2, 1), (2, 1, 1), (3, 1, 1), (3, 2, 1), (2.8, 2, 1), (2.8, 1.2, 1), (2.2, 1.2, 1), (2.2, 2, 1)

# matrix with row vectors of points
A = np.array([p, q, r, s, t, u, v, w, x, y, z, aa, ab, ac, ad])

# create the translation transformation matrix
# set x and y translation values here (x = ?, y = ?)
x = 4
y = 2
L = np.array([[1, 0, x], [0, 1, y], [0, 0, 1]])

# plot
fig = plt.figure()
ax = plt.gca()
ax.set_aspect(1)
xs = []
ys = []
for row in A:
    output_row = np.matmul(L, row)
    x, y, z = output_row
    xs.append(x)
    ys.append(y)
    plt.scatter(x, y)
xs.append(xs[0])
ys.append(ys[0])
plt.plot(xs, ys, color="black", linestyle='solid')
plt.xlim([0, (x + 1)])
plt.ylim([0, (y + 1)])
plt.show()
```
